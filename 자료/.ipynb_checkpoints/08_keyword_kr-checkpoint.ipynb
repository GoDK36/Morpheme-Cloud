```python
from konlpy.tag import Kkma  
import nltk
import re

tagger=Kkma() 
```

```python
import jpype
jpype.startJVM(jpype.getDefaultJVMPath())
```

```python
# 한국어 텍스트 불러 오기
rsc = r'D:\Programming\python\창회선배스터디\Morpheme_Cloud\자료\토지2.txt'

with open(rsc, 'r', encoding='utf8') as kr_f:
    txt = kr_f.read()
```

```python
def flat(content):
    return ["{}/{}".format(word, tag) for word, tag in tagger.pos(content)]

flat(txt[:30])
```

```python
# 한국어 텍스트 불러 오기
tagged_list = []
with open(rsc, 'r', encoding='utf8') as kr_f:
    for line in kr_f:
        line = line.strip()
        tagged_list += tagger.pos(line[:100])
print(tagged_list[:30])
```

```python
def is_stopword_kr(tag):
    if re.search('^[EXJS]', tag, re.M):
        return True
    else:
        return False
```

```python
# 내용어 리스트 생성
ctt_words = []
for morph, tag in tagged_list:
    if is_stopword_kr(tag):continue
    ctt_words.append(morph)
```

```python
ctt_words_nltk = nltk.Text(ctt_words)
```

```python
# 차트에 표시될 폰트 설정
import matplotlib.font_manager as fm
import matplotlib
#font_location = r"C:\Windows\Fonts\NotoSansCJKkr-Regular.otf"
font_location = r"C:\WINDOWS\Fonts\NGULIM.TTF"

font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
```

```python
# from matplotlib import pylab
# pylab.show = lambda: pylab.savefig('some_filename.png')
ctt_words_nltk.plot(30) # plot(N): 빈도 상위 50개의 단어에 대한 차트 생성
```

```python
txt_nltk = nltk.Text([m for m, t in tagged_list])
```

```python
print(txt_nltk.concordance('호랑이', width=100, lines=100))
```

```python
txt_nltk.dispersion_plot(['호랑이', '나무꾼'])
```

### 워드클라우드 생성
package: pip install wordcloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
```

```python
word_freq_d = defaultdict(int)
tag_match = re.compile('\S+[NVAD]S')
res = tag_match.findall(tagged_list)
print(res)
```

```python
word_freq_d = defaultdict(int)
# tag_match = re.compile('\S+[NVAD]S')
for morph, tag in tagged_list:
    if tag.startswith('N'):
        word_freq_d[morph] += 1
    elif tag.startswith('V'):
        word_freq_d[morph] += 1
    elif tag.startswith('A'):
        word_freq_d[morph] += 1
    elif tag.startswith('D'):
        word_freq_d[morph] += 1
print(word_freq_d)
```

```python
# 내용어 리스트 생성
nouns_freq_d = defaultdict(int)
for morph, tag in tagged_list:
    if tag.startswith('N'):
        nouns_freq_d[morph] += 1
print(nouns_freq_d)
```

```python
wordcloud = WordCloud(font_path=r"C:\WINDOWS\Fonts\NGULIM.TTF")
wordcloud = wordcloud.generate_from_frequencies(nouns_freq_d)
```

```python
fig = plt.figure(figsize=(12,12)) # 이미지 사이즈
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
